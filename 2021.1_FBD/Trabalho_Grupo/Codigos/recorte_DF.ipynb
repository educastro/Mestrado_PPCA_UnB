```python
import pandas as pd
```

```python
alunos = pd.read_csv("SUP_ALUNO_2019.csv", sep="|", encoding='ISO-8859-1')

alunos.head()
```

```python
ies = pd.read_csv("SUP_IES_2019.csv", sep="|", encoding='ISO-8859-1')

ies_df = ies[ies["CO_UF"] == 53]
ies_df_codigos = ies_df["CO_IES"].values.tolist()

ies_df.to_csv('SUP_IES_2019_DF.csv', sep=";", encoding='ISO-8859-1', index = False)
ies_df
```

```python
docentes = pd.read_csv("SUP_DOCENTE_2019.csv", sep="|", encoding='ISO-8859-1')

docentes_df = docentes[docentes["CO_IES"].isin(ies_df_codigos)]

docentes_df.to_csv('SUP_DOCENTE_2019_DF.csv', sep=";", encoding='ISO-8859-1', index = False)
docentes_df
```

```python
cursos = pd.read_csv("SUP_CURSO_2019.csv", sep="|", encoding='ISO-8859-1')

cursos_df = cursos[cursos["CO_IES"].isin(ies_df_codigos)]

cursos_df.to_csv('SUP_CURSO_2019_DF.csv', sep=";", encoding='ISO-8859-1', index = False)
cursos_df
```

```python
alunos_df = alunos[alunos["CO_IES"].isin(ies_df_codigos)]

alunos_df = alunos_df.drop(columns=['NU_ANO_CENSO', 'CO_CURSO_POLO', 'CO_CINE_ROTULO', 'CO_ALUNO_CURSO_ORIGEM', 'NU_IDADE', 'TP_NACIONALIDADE', 'CO_UF_NASCIMENTO', 'CO_MUNICIPIO_NASCIMENTO', 'IN_DEFICIENCIA_AUDITIVA', 'IN_DEFICIENCIA_FISICA', 'IN_DEFICIENCIA_INTELECTUAL', 'IN_DEFICIENCIA_MULTIPLA', 'IN_DEFICIENCIA_SURDEZ', 'IN_DEFICIENCIA_SURDOCEGUEIRA', 'IN_DEFICIENCIA_BAIXA_VISAO', 'IN_DEFICIENCIA_CEGUEIRA', 'IN_DEFICIENCIA_SUPERDOTACAO', 'IN_TGD_AUTISMO', 'IN_TGD_SINDROME_ASPERGER', 'IN_TGD_SINDROME_RETT', 'IN_TGD_TRANSTOR_DESINTEGRATIVO', 'IN_INGRESSO_VESTIBULAR', 'IN_INGRESSO_ENEM', 'IN_INGRESSO_AVALIACAO_SERIADA', 'IN_INGRESSO_SELECAO_SIMPLIFICA', 'IN_INGRESSO_OUTRO_TIPO_SELECAO', 'IN_INGRESSO_VAGA_REMANESC', 'IN_INGRESSO_VAGA_PROG_ESPECIAL', 'IN_INGRESSO_TRANSF_EXOFFICIO', 'IN_INGRESSO_DECISAO_JUDICIAL', 'IN_INGRESSO_CONVENIO_PECG', 'IN_INGRESSO_EGRESSO', 'IN_INGRESSO_OUTRA_FORMA', 'IN_RESERVA_VAGAS', 'IN_RESERVA_ETNICO', 'IN_RESERVA_DEFICIENCIA', 'IN_RESERVA_ENSINO_PUBLICO', 'IN_RESERVA_RENDA_FAMILIAR', 'IN_RESERVA_OUTRA', 'IN_FIN_REEMB_FIES', 'IN_FIN_REEMB_ESTADUAL', 'IN_FIN_REEMB_MUNICIPAL', 'IN_FIN_REEMB_PROG_IES', 'IN_FIN_REEMB_ENT_EXTERNA', 'IN_FIN_REEMB_OUTRA', 'IN_FIN_NAOREEMB_PROUNI_INTEGR', 'IN_FIN_NAOREEMB_PROUNI_PARCIAL', 'IN_FIN_NAOREEMB_ESTADUAL', 'IN_FIN_NAOREEMB_MUNICIPAL', 'IN_FIN_NAOREEMB_PROG_IES', 'IN_FIN_NAOREEMB_ENT_EXTERNA', 'IN_FIN_NAOREEMB_OUTRA', 'IN_INGRESSO_VAGA_NOVA', 'IN_INGRESSO_PROCESSO_SELETIVO'])
alunos_df = alunos_df.fillna(0)

alunos_df.to_csv('SUP_ALUNO_2019_DF.csv', sep=";", encoding='ISO-8859-1', index = False)
```
