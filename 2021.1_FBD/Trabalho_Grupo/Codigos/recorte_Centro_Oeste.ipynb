```python
import pandas as pd
```

```python
alunos = pd.read_csv("SUP_ALUNO_2019.csv", sep="|", encoding='ISO-8859-1')
```

```python
ies = pd.read_csv("SUP_IES_2019.csv", sep="|", encoding='ISO-8859-1')

ies_df = ies[ies["CO_REGIAO"] == 5]
ies_df_codigos = ies_df["CO_IES"].values.tolist()

ies_df.to_csv('SUP_IES_2019_CENTRO_OESTE_com_indice.csv', sep=";", encoding='ISO-8859-1', index = True)
ies_df.to_csv('SUP_IES_2019_CENTRO_OESTE_sem_indice.csv', sep=";", encoding='ISO-8859-1', index = False)
ies_df
```

```python
docentes = pd.read_csv("SUP_DOCENTE_2019.csv", sep="|", encoding='ISO-8859-1')

docentes_df = docentes[docentes["CO_IES"].isin(ies_df_codigos)]

docentes_df.to_csv('SUP_DOCENTE_2019_CENTRO_OESTE_com_indice.csv', sep=";", encoding='ISO-8859-1', index = True)
docentes_df.to_csv('SUP_DOCENTE_2019_CENTRO_OESTE_sem_indice.csv', sep=";", encoding='ISO-8859-1', index = False)
docentes_df
```

```python
cursos = pd.read_csv("SUP_CURSO_2019.csv", sep="|", encoding='ISO-8859-1')

cursos_df = cursos[cursos["CO_IES"].isin(ies_df_codigos)]

cursos_df.to_csv('SUP_CURSO_2019_CENTRO_OESTE_com_indice.csv', sep=";", encoding='ISO-8859-1', index = True)
cursos_df.to_csv('SUP_CURSO_2019_CENTRO_OESTE_sem_indice.csv', sep=";", encoding='ISO-8859-1', index = False)
cursos_df
```

```python
alunos_df = alunos[alunos["CO_IES"].isin(ies_df_codigos)]

alunos_df.to_csv('SUP_ALUNO_2019_CENTRO_OESTE_com_indice.csv', sep=";", encoding='ISO-8859-1', index = True)
alunos_df.to_csv('SUP_ALUNO_2019_CENTRO_OESTE_sem_indice.csv', sep=";", encoding='ISO-8859-1', index = False)
alunos_df
```
