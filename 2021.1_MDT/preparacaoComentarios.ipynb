# LeIA é uma adaptação do VADER para o português
# https://github.com/rafjaa/LeIA.git

O valor <code>compound</code> pode ser utilizado para descrever o sentimento predominante no texto, por meio dos limites de valores:

- Sentimento positivo: <code>compound >= 0.05</code>
- Sentimento negativo: <code>compound <= -0.05</code>
- Sentimento neutro: <code>(compound > -0.05) and (compound < 0.05)</code>

```python
#!pip install vaderSentiment (versão em inglês)

from leia import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(str(sentence))
    #print("{:-<40} {}".format(sentence, str(score)))
    return(score)

def vader_compound(sentence):
    score = sentiment_analyzer_scores(sentence)
    return(score['compound'])

def vader_pos(sentence):
    score = sentiment_analyzer_scores(sentence)
    return(score['pos'])

def vader_sentiment(sentence):
    score = sentiment_analyzer_scores(sentence)
    sentiment = 'negativo'
    if(score['compound']>=0.05):
        sentiment = 'positivo'
    elif(score['compound']> -0.05 and score['compound'] < 0.05):
        sentiment = 'neutro'
    return(sentiment)
```

```python
# exemplo de uma avaliação de classificou com 5 estrelas
txt='''
CADA VEZ O APP ESTA MELHORANDO NOTA DEZ GOSTARIA DE DEIXA UMA IDEIA PODERIA VCS TAMBEM NESSE APP O SALDO DO FGTS O MAS 
EU NAO TENHO O QUE QUESTIONAR'''
print(sentiment_analyzer_scores(txt))
```

```python
txt='''Ótima ferramenta, ajudou muito na minha situação, ainda da para acompanhar todos os registros que tive em carteira!'''
print(sentiment_analyzer_scores(txt))
```
