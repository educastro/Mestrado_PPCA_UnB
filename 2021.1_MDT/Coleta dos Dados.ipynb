```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from google_play_scraper import Sort, reviews_all
import csv
```

```python
url = 'https://www.gov.br/pt-br/apps/@@galeria-de-aplicativos?b_start:int='

links_apps = []
paginador = 0

nome_arquivo = 'comentarios.csv'
```

```python
for iterator in range (1,9):
    url_get = url + str(paginador)
    
    html = urlopen(url_get)
    bs = BeautifulSoup(html, 'html.parser')
    
    apps = bs.find_all('a', {'class': 'android'})
    
    for app in apps:
        itens = app.get('href')
        
        print(itens)
        print("-"*50)
        links_apps.append(itens)
        
    paginador += 30
    print("="*50)
```

```python
links_apps.remove("https://play.google.com/store/apps/details?id=br.gov.beta.ajudaauditor")
# links_apps.remove("https://play.google.com/store/apps/details?id=br.gov.caixa.auxilio&hl=pt_BR")
# links_apps.remove("https://play.google.com/store/apps/details?id=br.gov.dataprev.carteiradigital")
# links_apps.remove("https://play.google.com/store/apps/details?id=br.gov.serpro.cnhe")
```

```python
with open(nome_arquivo, 'w', newline="", encoding='utf-8') as csvfile: 
    csv_writer = csv.writer(csvfile, delimiter=';')
    header = ['app', 'userName', 'content', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt']
    csv_writer.writerow(header)
    
    for link in links_apps:

        nome_app = link.split("id=")[1]
        nome_app = nome_app.split("&hl=")[0]
        print(nome_app)
        
        print("Iniciando a coleta...")

        result = reviews_all(
            nome_app,
            lang='pt-br',# defaults to 'en'
            country='br', # defaults to 'ke'
            sort=Sort.NEWEST
        )

        print("Finalizando a coleta...")
        print("Iniciando o tratamento...")
             
        for element in result:
            row = []
            row.append(nome_app)
            
            row.append(element["userName"])
            row.append(element["content"])
            row.append(element["score"])
            row.append(element["thumbsUpCount"])
            row.append(element["reviewCreatedVersion"])
            row.append(str(element["at"])[:10])
            row.append(element["replyContent"])
            row.append(str(element["repliedAt"])[:10])
            csv_writer.writerow(row)
            #print(row)
            
        
        print("Finalizando o tratamento...")
        print("-"*30)
    
    
```
